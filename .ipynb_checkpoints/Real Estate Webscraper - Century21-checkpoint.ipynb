```python
# import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# save URL of interest
URL_oi = "https://www.century21.com/real-estate/brooklyn-ny/LCNYBROOKLYN/?sa=CNYASTORIA%2CCNYBREEZYPOINT%2CCNYELMHURST%2CCNYHOWARDBEACH%2CCNYJACKSONHEIGHTS%2CCNYLONGISLANDCITY%2CCNYMASPETH%2CCNYMIDDLEVILLAGE%2CCNYOZONEPARK%2CCNYREGOPARK%2CCNYRIDGEWOOD%2CCNYSUNNYSIDE%2CCNYWOODHAVEN%2CCNYWOODSIDE&pr=%7C350000"
# load page
r = requests.get(URL_oi)
r
```

```python
# get content from returned request
c = r.content
# c
```

```python
# parse using BeautifulSoup
soup = BeautifulSoup(c, "html.parser")
```

```python
# # print out parsed text (with better readability)
# soup.body.prettify
```

```python
# get property card info div text
properties = soup.find_all("div", {"class":"property-card-primary-info"})
properties[0]
```

```python
# iterate over properties and print address and price info
for prop in properties:
    address = prop.find_all("div",{"class":"property-address"}) #NOTE: returns a list
    address = address[0].text #get first (and likely only) item from list and print only text within the tags
    address = address.strip() #remove leading and trailing whitespace
    price = prop.find("a", {"class":"listing-price"}).text.strip() 
    try: #account for missing bedroom number info
        beds = prop.find("div", {"class":"property-beds"}).strong.text.strip() #get number of bed roows in string-tag
    except:
        beds = "not specified"
    try:
        size = str(prop.find("div", {"class":"property-sqft"}).strong.text.strip())+" sqft."
    except:
        size = "not specified"
    # print information per entry
    print(f"{address}: {price}\n  Bedrooms: {beds}\n  Size: {size}\n\n")
```

<u>NOTE</u>: Using __.find()__ instead of __.find_all()__ returns sinlge tag entry instead of a list <br>
-> _[0]_ does not have to be specified anymore <br>
--> use for extractions where only one entry is expected

```python
# create df to store information in 
df = pd.DataFrame()

# iterate over properties, save overview info to df; go to property-specific URL and also add specific property info to df
for prop in properties:
    # create dict to save information to pass to df
    d ={}
    
    ##Overview data
    # save property-overview info to variables
    address = prop.find_all("div",{"class":"property-address"}) #NOTE: returns a list
    address = address[0].text #get first (and likely only) item from list and print only text within the tags
    address = address.strip() #remove leading and trailing whitespace
    price = prop.find("a", {"class":"listing-price"}).text.strip() 
    try: #account for missing bedroom number info
        beds = prop.find("div", {"class":"property-beds"}).strong.text.strip() #get number of bed roows in string-tag
    except:
        beds = "not specified"
    try:
        size = str(prop.find("div", {"class":"property-sqft"}).strong.text.strip())+" sqft."
    except:
        size = "not specified"
        
    # add overview info to dict via their variable-names
    d['Address'] = address
    d['Price'] = price
    d['Bedrooms'] = beds
    d['Size'] = size
    
    ##Property-specific Data
    # get URL for property-specific website
    URL_addition = prop.find("a",{"class":"listing-price"}, href=True)["href"]
    # save complete website-URL for property
    prop_URL= "https://www.century21.com" + URL_addition
    
    # get property-website content
    r_prop = requests.get(prop_URL, headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
    c_prop = r_prop.content
    prop_spec = BeautifulSoup(c_prop, "html.parser")
    
    # get property features info and save in variable
    features = prop_spec.find("div", {"id":"property-features"}).find_all("li")
    # iterate over property features and add features and values to dict
    f = {}
    for feat in features:
        feat_name, feat_value = feat.text.split(":")
        d[feat_name] = feat_value
    
    ##save dict information in df
    df = df.append(d, ignore_index=True)
```
